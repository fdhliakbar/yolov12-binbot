# YOLOv11

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
# Menggunakan model YOLOv8 untuk deteksi objek dengan model pratrained. n berarti nano , versi paling ringan.
model = YOLO('yolov8n.pt')
```

```python
results = model('../data/images/Dog.jpg')

# Tampilkan hasil
for r in results:
    im_array = r.plot() 
    plt.imshow(im_array)
    plt.axis('off')
    plt.show()
```
